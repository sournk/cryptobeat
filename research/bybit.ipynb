```python
import requests

# Конфигурация API
symbol = 'BTCUSDT'  # Символ фьючерса

# URL для запроса цены фьючерсов
url = f"https://api.bybit.com/v2/public/tickers?symbol={symbol}"

# Выполнение запроса
response = requests.get(url)

# Проверка успешного выполнения запроса
if response.status_code == 200:
    data = response.json()
    if data['ret_code'] == 0:
        price = data['result'][0]['last_price']
        print(f"Текущая цена фьючерса {symbol}: {price}")
    else:
        print(f"Ошибка при получении цены фьючерса: {data['ret_msg']}")
else:
    print("Ошибка при выполнении запроса")
```

```python
import requests


def get_historical_candles(symbol, interval, limit):
    url = f"https://api.bybit.com/v2/public/kline/list?symbol={symbol}&interval={interval}&limit={limit}&from=1621600000"
    response = requests.get(url)
    data = response.json()

    if data['ret_code'] == 0:
        candles = data['result']
        return candles
    else:
        print("Ошибка при получении свечей:", data['ret_msg'])
        return []


symbol = "BTCUSDT"  # Символ BTCUSDT на Bybit
# Минутный интервал (1 - 1 минута, 3 - 3 минуты, 5 - 5 минут и т.д.)
interval = "1"
limit = 100  # Количество свечей, которые нужно получить

candles = get_historical_candles(symbol, interval, limit)

# Проверка на наличие данных
if candles:
    # Обработка полученных свечей
    for candle in candles:
        timestamp = candle['open_time']  # Временная метка
        open_price = candle['open']  # Цена открытия
        high_price = candle['high']  # Максимальная цена
        low_price = candle['low']  # Минимальная цена
        close_price = candle['close']  # Цена закрытия
        volume = candle['volume']  # Объем

        print(f"Временная метка: {timestamp}, Открытие: {open_price}, "
              f"Максимум: {high_price}, Минимум: {low_price}, "
              f"Закрытие: {close_price}, Объем: {volume}")
else:
    print("Нет данных о свечах.")
```

```python
import secret
from pybit.unified_trading import HTTP

session = HTTP(testnet=False, api_key=secret.BYBIT_API_KEY,
               api_secret=secret.BYBIT_API_KEY)
res = session.get_instruments_info(
    category="linear",
    # symbol="BTCUSDT",
)

symbols = res['result']['list']

l = []
for s in symbols:
    price_tick = s['priceFilter']['tickSize']
    qty_step = s['lotSizeFilter']['qtyStep']
    l.append({'symbol': s['symbol'],
              'priceTick': price_tick,
              'qtyStep': qty_step})

print('TOP3 smallest priceTick')
l = sorted(l, key=lambda x: x['priceTick'])
for s in l[:3]:
    print(f"{s['symbol']=} {s['priceTick']=} {s['qtyStep']=}")

print('TOP3 smallest qtyStep')
l = sorted(l, key=lambda x: x['qtyStep'])
for s in l[:3]:
    print(f"{s['symbol']=} {s['priceTick']=} {s['qtyStep']=}")
```
